```python
from pickle import FALSE
import re

################################################################################
class Gene:
  def __init__(self,name):
    self.name = name
    self.exons = []
    self.cds = []
    self.mrna = []



  def setSequence(self,sequence):
    self.sequence = sequence

  def setExons(self,exon):
    self.exons.append((exon,len(self.exons)+1))

  def setCDS(self,sequence):
    self.cds.append((sequence,len(self.cds)+1))

  def setMRNA(self,mrnaSequence):
    self.mrna.append((mrnaSequence,len(self.mrna)+1))
################################################################################
class E1:
  def __init__(self,name):
    self.name = name

#------------------------ return string

  def _getStringFromFASTA(self,customList):
    indice_primo_invio = customList.find('\n')
    customList = customList[indice_primo_invio + 1:]
    customString = ''.join(customList).replace('\n', '').replace(' ', '')
    return customString

#------------------------ return start and end indexes

  def _getIndexesInBiggerOne(searched, bigger):
    start_index = bigger.find(searched)
    if start_index != -1:
        end_index = start_index + len(searched) - 1
        return start_index, end_index
    else:
        return None



#________________________ length

  def returnLength(self,gene,printerBOOL):
    if isinstance(gene, Gene) and isinstance(printerBOOL,bool):
      if(printerBOOL):
        print("-/// E1 class method called. \\\\\\- \n", "--- method 'returnLength' running ... ---")
        print(self._getStringFromFASTA(gene.sequence))
        print("Length of " + gene.name +" is " + str(len(self._getStringFromFASTA(gene.sequence))))
      else:
        return len(self._getStringFromFASTA(gene.sequence))
    else:
      raise Exception(" !- The first variable is not an istance of Gene class or the second one is not a bool. -!")

#________________________ GC amount

  def GCamount(self,gene,printerBOOL):
    if isinstance(gene, Gene) and isinstance(printerBOOL,bool):
      M = re.finditer(r"GC",self._getStringFromFASTA(gene.sequence))
      if M:
        if(printerBOOL):
          print("-/// E1 class method called. \\\\\\- \n", "--- method 'GCamount' running ... ---")
          print("The amount of 'GC' pattern in this sequence is " + str(len(list(M))*2 / self.returnLength(gene,False)) + "%")
        else:
          return len(list(M))*2 / self.returnLength(gene,False)
      else:
        print("... 'GC' pattern not found ...")
    else:
      raise Exception(" !- The first variable is not an istance of Gene class or the second one is not a bool. -!")

#________________________ valid CDS

  def CDSchecker(self,gene,printerBOOL):
    if isinstance(gene, Gene) and isinstance(printerBOOL,bool):
        if(printerBOOL):
          print("-/// E1 class method called. \\\\\\- \n", "--- method 'CDSchecker' running ... ---")
          for idx, cds in enumerate(gene.cds, start=1):

              M = re.finditer(r"^ATG(:?A|T|G|C)+TGA$",cds[0])
              if M:
                print(f"CDS {idx} valid.")
              else:
                print(f"!CDS {idx} invalid.!")
    else:
      raise Exception(" !- The first variable is not an istance of Gene class or the second one is not a bool. -!")

  #________________________ reverse complement calculator

  def reverseComplement(self,gene,printerBOOL):
    if isinstance(gene, Gene) and isinstance(printerBOOL,bool):
        complement_dict = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
        reverse_comp = ''.join(complement_dict[base] for base in reversed(self._getStringFromFASTA(gene.sequence)))
        if(printerBOOL):
          print("-/// E1 class method called. \\\\\\- \n", "--- method 'reverseComplement' running ... ---")
          chops = [reverse_comp[i:i+70] for i in range(0, len(reverse_comp), 70)]
          reverse_comp = '\n'.join(chops)
          print(reverse_comp)
        else:
          return reverse_comp
    else:
      raise Exception(" !- The first variable is not an istance of Gene class or the second one is not a bool. -!")
#________________________ ATG count

  def ATGcount(self,gene,printerBOOL):
    if isinstance(gene, Gene) and isinstance(printerBOOL,bool):
      M = re.finditer(r"ATG",self._getStringFromFASTA(gene.sequence))
      if M:
        if(printerBOOL):
          print("-/// E1 class method called. \\\\\\- \n", "--- method 'ATGcount' running ... ---")
          print("The count of 'ATG' pattern in this sequence is " + str(len(list(M))) + ".")
        else:
          return len(list(M))
      else:
        print("... 'ATG' pattern not found ...")
    else:
      raise Exception(" !- The first variable is not an istance of Gene class or the second one is not a bool. -!")


  #________________________ repetitions finder

  def areThereAnyReps(self,gene,atLeastOrExactlyN,greedy,printerBOOL):
    if isinstance(gene, Gene) and isinstance(printerBOOL,bool) and isinstance(atLeastOrExactlyN,int) and isinstance(greedy,bool):
      if(greedy):
        regex_pattern = re.compile(f"(A){{{atLeastOrExactlyN},}}|(T){{{atLeastOrExactlyN},}}|(C){{{atLeastOrExactlyN},}}|(G){{{atLeastOrExactlyN},}}")
      else:
        regex_pattern = re.compile(f"((A){{{atLeastOrExactlyN},}}?|(T){{{atLeastOrExactlyN},}}?|(C){{{atLeastOrExactlyN},}}?|(G){{{atLeastOrExactlyN},}}?")

      M = re.finditer(regex_pattern,self._getStringFromFASTA(gene.sequence))
      if M:
        if(printerBOOL):
          print("-/// E1 class method called. \\\\\\- \n", "--- method 'areThereAnyReps' running ... ---")
          print(f"The count of {atLeastOrExactlyN} reps pattern in this sequence is " + str(len(list(M))) + ".")
        else:
          return len(list(M))
      else:
        print("... Reps not found ...")
    else:
      raise Exception(" !- Check the nature of the input variables. -!")

  #________________________ polymerase

  def polymerase(self,gene,printerBOOL):
      code = {
          "TTT": "F", "TTC": "F", "TTA": "L", "TTG": "L",
          "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S",
          "TAT": "Y", "TAC": "Y", "TAA": "*", "TAG": "*",
          "TGT": "C", "TGC": "C", "TGA": "*", "TGG": "W",
          "CTT": "L", "CTC": "L", "CTA": "L", "CTG": "L",
          "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P",
          "CAT": "H", "CAC": "H", "CAA": "Q", "CAG": "Q",
          "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R",
          "ATT": "I", "ATC": "I", "ATA": "I", "ATG": "M",
          "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T",
          "AAT": "N", "AAC": "N", "AAA": "K", "AAG": "K",
          "AGT": "S", "AGC": "S", "AGA": "R", "AGG": "R",
          "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",
          "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A",
          "GAT": "D", "GAC": "D", "GAA": "E", "GAG": "E",
          "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G",
      }
      if isinstance(gene, Gene) and isinstance(printerBOOL,bool):
        for cds in gene.cds:
          sequence = ""
          cds = self._getStringFromFASTA(cds[0])
          for i in range(0, len(cds), 3):
            codone = cds[i:i+3]
            aminoacido = code.get(codone, "X")  # "X" se il codone non è nel dizionario
            sequence += aminoacido

          print(sequence + "\n")
      else:
        raise Exception(" !- Check the nature of the input variables. -!")


  #________________________ search an ORF

  def ORFfinder(self,gene,greedy,printerBOOL):
      if isinstance(gene, Gene) and isinstance(printerBOOL,bool) and isinstance(greedy,bool):
          if(printerBOOL):
              print("-/// E1 class method called. \\\\\\- \n", "--- method 'ORFfinder' running ... ---")
          i = 0
          cdsListString = []

          for cds in gene.cds:
            cdsListString.append(self._getStringFromFASTA(cds[0]))

          listsOrfMRNAVariants = []
          for mrna in gene.mrna:
            i = i+1

            if(printerBOOL):
              print(">>>>>>>>>> RNA TRASCRIPT VARIANT n. ", i, " <<<<<<<<<<")
            mrna = self._getStringFromFASTA(mrna[0])

            if(greedy):
              patternRegex = re.compile(r'ATG(?:[ACGT]{3})+(?:TAA|TAG|TGA)')
            else:
              patternRegex = re.compile(r'ATG(?:[ACGT]{3})+?(?:TAA|TAG|TGA)')

            matches = patternRegex.finditer(mrna)

            openReadingFrames = []

            for match in matches:
              orf = match.group()
              indice_start_orf = match.start()
              if(printerBOOL):
                if(orf in cdsListString):
                  print("!-- CDS FIND --!:",orf)
                else:
                  print("-- ORF FIND --:", orf)

                print(" start at ", indice_start_orf)
              openReadingFrames.append((orf, indice_start_orf))

            listsOrfMRNAVariants.append((openReadingFrames,i))
          if(not printerBOOL):
            return listsOrfMRNAVariants

      else:
          raise Exception(" !- Check the nature of the input variables. -!")


  #________________________ get introns

  def getIntron(self, gene,printerBOOL):
      if isinstance(gene, Gene) and isinstance(printerBOOL,bool):
        if(printerBOOL):
              print("-/// E1 class method called. \\\\\\- \n", "--- method 'getIntron' running ... ---")
        exonsToChop = []
        for exon in gene.exons:
          exonsToChop.append(self._getStringFromFASTA(exon[0]))
        pattern = '|'.join(map(re.escape, exonsToChop))
        introns = re.split(pattern, self._getStringFromFASTA(gene.sequence))
        introns = [seq for seq in introns if seq]
        if(printerBOOL):
          for intr in introns:
            print("-----------------------------")
            print(intr)
        else:
          return introns
      else:
          raise Exception(" !- Check the nature of the input variables. -!")



################################################################################



import pandas as pd

def readFASTA(file_path):
    # with Open function it doesn't work on GitHub, so we read the file as a csv
    df = pd.read_csv(file_path, sep='\t', header=None)
    # Convert dataframe to string
    file_contents = df.to_string(index=False, header=False)
    # List from lines
    lines = file_contents.split('\n')
    # Remove spaces
    modified_lines = [lines[0]] + [line.replace(' ', '') for line in lines[1:]]
    # Join the lines
    result_string = '\n'.join(modified_lines)
    return result_string


output_string = readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/gene.fasta")
# print(output_string)

esercitazione1 = E1("esercitazione1")

NANOG = Gene("NANOG")
NANOG.setSequence(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/gene.fasta"))
NANOG.setCDS(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/t1CDS.fasta"))
NANOG.setCDS(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/t2CDS.fasta"))
NANOG.setExons(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/Exons/1.fasta"))
NANOG.setExons(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/Exons/2.fasta"))
NANOG.setExons(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/Exons/3.fasta"))
NANOG.setExons(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/Exons/4.fasta"))
NANOG.setExons(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/Exons/5.fasta"))

NANOG.setMRNA(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/t1.fasta"))
NANOG.setMRNA(readFASTA("https://raw.githubusercontent.com/simonc999/Bioinformatics_python_project/main/eukaryotes/nanog/t2.fasta"))




esercitazione1.returnLength(NANOG,True)
esercitazione1.GCamount(NANOG,True)
esercitazione1.CDSchecker(NANOG,True)
rc = esercitazione1.reverseComplement(NANOG,False)
esercitazione1.ATGcount(NANOG,True)
esercitazione1.areThereAnyReps(NANOG,3,True,True)
esercitazione1.polymerase(NANOG,True)
esercitazione1.ORFfinder(NANOG,False,True)
esercitazione1.getIntron(NANOG,True)
# mancano 4 10 12 13
```
